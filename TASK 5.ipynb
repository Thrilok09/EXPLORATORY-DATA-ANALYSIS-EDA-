```python
plt.hist(df['PassengerId'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Passenger IDs')
plt.xlabel('PassengerId')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(df['Survived'], bins=2, color='lightgreen', edgecolor='black', rwidth=0.8)
plt.title('Survival Distribution')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.xticks([0, 1])
plt.show()
```
